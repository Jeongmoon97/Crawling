```python
from selenium import webdriver
import time
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import random
import pandas as pd
```

```python
browser = webdriver.Chrome("./chromedriver")
```

```python
# url 저장
def move_page(searching, page, start, end):
    url = f'https://section.blog.naver.com/Search/Post.naver?pageNo={page}&rangeType=PERIOD&orderBy=sim&startDate={start}&endDate={end}&keyword={searching}'
    return url
```

```python
searching = 'ebs 온라인 클래스'
start = '2020-07-16'
end = '2021-05-31'
result = []
```

```python
for page in tqdm(range(1,573)):
    url = move_page(searching, page, start, end)
    browser.get(url)
    time.sleep(2)
    
    article_list = browser.find_elements_by_css_selector('a.desc_inner')
    
    date=[]
    for i in range(len(article_list)):
        date_list = browser.find_elements_by_css_selector('span.date')
        date.append(date_list[i].text)
    
    # 글 선택
    contents = []
    for article in article_list:
        
        article.click()
        time.sleep(2)
        
        # 탭 변경
        change_tab = browser.window_handles[-1]
        browser.switch_to.window(change_tab)
        
        try:
            browser.switch_to.frame('mainFrame')
            html = browser.page_source
            soup = BeautifulSoup(html,'html.parser')
            try:
                content = soup.select('div.se-main-container')[0].text.strip()
            except:
                pass
            contents.append(content)
        except:
            pass
                
        time.sleep(0.5)
        browser.close()
        
        change_tab = browser.window_handles[-1]
        browser.switch_to.window(change_tab)
    
    for i in range(len(article_list)):
        data = [date[i],contents[i]]
        result.append(data)
    
    time.sleep(0.5)
```

```python
df = pd.DataFrame(result, columns=['date','content'])
df
```
