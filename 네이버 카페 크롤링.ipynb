```python
from selenium import webdriver
import time
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import random
import pandas as pd
```

```python
browser = webdriver.Chrome("./chromedriver")
```

```python
# url 저장
def move_page(searching, page):
    url = f'https://section.cafe.naver.com/ca-fe/home/search/articles?q={searching}&p={page}'
    return url
```

```python
searching = '비대면 강의'
result = []
```

```python
def get_data(browser):
    
    # 카페 프레임 변경
    browser.switch_to.frame('cafe_main')
    
    html = browser.page_source
    soup = BeautifulSoup(html,'html.parser')
    
    title = soup.select('h3.title_text')[0].text.strip()
    date = soup.select('span.date')[0].text[:10]
    
    # 카페명
    browser.switch_to.default_content()
    name = soup.find_all(attrs={'class':'footer', 'class':'cafe_name'})
    
    try:
        content = soup.select('div.se-module')[0].text.strip()
    except:
        pass
    
    all_reply = soup.select('div.comment_text_box')
    
    reply_list=[]
    
    for reply in all_reply:
        
        reply = reply.text.strip()
        reply_list.append(reply)
        
    separate_mark = '|||||'
    reply_joined = separate_mark.join(reply_list)
    
    data = [title, date, content, reply_joined]
    
    return data
```

```python
for page in range(1,3):
    
    url = move_page(searching, page)
    
    browser.get(url)
    time.sleep(2)
    
    article_list = browser.find_elements_by_css_selector('a.item_subject')
    
    # 글 선택
    for article in article_list:
        
        article.click()
        time.sleep(2)
        
        # 탭 변경
        change_tab = browser.window_handles[-1]
        browser.switch_to.window(change_tab)
        
        try:
            data = get_data(browser)    
            result.append(data)
            
        except:
            pass
        
        browser.close()
        time.sleep(1)
        
        change_tab = browser.window_handles[-1]
        browser.switch_to.window(change_tab)
        time.sleep(1)
```
