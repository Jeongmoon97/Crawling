## 파이썬으로 파일 만들기

1. open 함수는 파일을 원하는 형식으로 열어줍니다. 이 때 변수는 열어준 파일을 의미합니다.

```python
파일변수 = open( '파일이름', '오픈형식' )
```

- w (write) : 파일을 새로 쓸 때
- r (read) : 이미 존재하는 파일에서 데이터를 읽을 때
- a (append) : 이미 존재하는 파일에 내용을 이어 쓰고 싶을 때


2. close 함수는 파일변수가 가리키는 파일을 닫아줍니다. 파일을 닫아주지 않으면 프로그램이 종료되더라도 파일이 열려있는 채로 남아있을 수 있습니다.

```python
파일변수.close()
```


3. write 함수를 사용하면 덱스트 데이터를 파일변수가 가리키는 파일에 쓸 수 있습니다. 줄바뚬이 표현되어있지 않기 때문에 줄바꿈을 원하는 경우 \n을 추가합니다. (n은 new line의 약자입니다. \을 활용한 다양한 문장부호 표현법을 알고 싶다면 escape code를 공부하세요.)

```python
파일변수.write('텍스트 데이터')
```

## 파이썬으로 csv 파일 만들기

- csv는 comma seperated value의 약자로, 콤마로 데이터를 구분해서 저장하는 방법입니다. 이 때 콤마를 사용하면 데이터의 열이 구분되고, 줄바꿈을 사용하면 데이터의 열이 구분됩니다.

```python
f = open('hitsong.csv', 'w', encoding = 'UTF-8')
singers = ['박정현', '임창정', 'izi', '아이유']
songs = ['꿈에', '소주한잔', '응급실', '좋은날']
for i in range(len(singers)):
    f.write(singers[i] + ',' + songs[i] + '\n')
f.clsoe()
```

- 컴퓨터가 한글을 이해하는 방식이 다른 경우, 글자 깨짐 현상이 발생하는데요, encoding 옵션을 사용하면 이해하는 방식을 구체적으로 지정해주기 때문에 글자 깨짐현상을 피할 수 있습니다.
```python
f = open('hitsong.csv', 'w', encoding = 'UTF-8')
```

## 단계

1. 파일만들고 헤더 추가하기
2. 파일에 데이터 입력하기
3. 데이터 갈무리하기

### ex1. 네이버 tv

````python
import requests
from bs4 import BeautifulSoup

# 1단계
# 1-1. tv.naver.csv파일을 쓰기모드(w)로 열어주기
f = open('tv.naver.csv','w', encoding = 'UTF-8')

# 1-2. 헤더 추가하기
f.write('제목, 체널명, 조회수, 좋아요수\n')

raw = requests.get('http://tv.naver.com/r')
html = BeautifulSoup(raw.text, 'html.parser')

container = html.select('dl.cds_info')

for clip in container:
    title = clip.select_one('dt.title').text.strip()
    chn = clip.select_one('dd.chn').text.strip()
    hit = clip.select_one('span.hit').text.strip()
    like = clip.select_one('span.like').text.strip()
    # 3단계. replace 함수와 슬라이싱을 이용하여 데이터 갈무리하기
    # 3-1. ',' 제가하여 열 구분 없애기
    # hit/like는 숫자로 만드는 게 좋으므로 '_'가 아니라 삭제
    title = title.replace(',','_')
    chn = chn.replace(',', '_')
    hit = hit.replace(',', '')
    like = like.replace(',','')
    # 3-2. '재생 수', '좋아요 수' 글자 없애기
    hit = hit.replace('재생 수','')
    like = like[5:]
    print(title)
    print(chn)
    print(hit)
    print(like)
    print("="*50)
    # 2단계. 파일에 내용을 입력하기
    f.write(title + ',' + chn + ',' + hit + ',' + like + '\n')

# 1단계. 작업이 끝난 파일을 반복문 밖에서 닫아주기
f.close()
````

### ex2. 네이버 news

````python
import requests
from bs4 import BeautifulSoup

f=open('news.naver.csv','w', encoding = 'UTF-8')

f.write('제목,출처\n')

for n in range(1, 100, 10):
    raw = requests.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query=코알라&start="+str(n),
                   headers={'User-Agent' : 'Morzilla/5.0'})
    html = BeautifulSoup(raw.text, "html.parser")
    
    articles = html.select("ul.type01>li")
    
    for ar in articles:
        title = ar.select_one("a._sp_each_title").text.strip().replace(',','_')
        source = ar.select_one("span._sp_each_source").text.strip().replace(',','_')
        print(title, source)
        f.write(title + ',' + source + '\n')

f.close()
````

### ex3. 네이버 ebook

````python
import requests
from bs4 import BeautifulSoup

f=open('ebook.naver.csv','w')
f.write('제목, 저자\n')

for i in range(1,6):
    raw= requests.get("https://series.naver.com/ebook/top100List.nhn?page="+str(i),
                 headers = {'User-Agent': 'Morzilla/5.0'})
    html = BeautifulSoup(raw.text,'html.parser')

    container = html.select("ul.lst_thum.v1>li")

    for cont in container:
        title = cont.select_one("a strong").text.strip()
        writer = cont.select_one("span.writer").text.strip()
        title = title.replace(',','_')
        writer = writer.replace(',', '_')
        f.write(title+','+writer+'\n')
        
f.close()
````
