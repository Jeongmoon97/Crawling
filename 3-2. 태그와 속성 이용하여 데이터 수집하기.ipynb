### 태그와 속성
- 태그는 html문서에서 기능을 정의하고, 속성은 태그를 사용할 때 태그의 성격, 목표 등을 설명하는 역할을 합니다.
- 예를 들어 a 태그의 href 속성은 연결되는 페이지의 주소를 저장하고, img 태그의 src 속성은 이미지 주소를 저장합니다.

### 링크 안에서 데이터 수집하기

#### 속성값 가져오기

```python
변수.attrs['속성이름']
```
- 선택한 코드에서 원하는 속성의 속성값을 저장합니다.
- title에 저장되어있는 a 태그의 href 값을 저장하려면 title.attrs['href']와 같이 쓸 수 있습니다.
- 또한 정확한 연결 주소를 만들기 위해서 href 값에 생략된 'http://movie.naver.com'을 붙여줍니다.

#### 새로운 페이지에서 데이터 수집하기
수집한 링크(url)에서 requests와 bs4를 통해 데이터 수집을 반복합니다.

````python
import requests
from bs4 import BeautifulSoup
raw = requests.get("https://movie.naver.com/movie/running/current.nhn",
                  headers={'User-Agent': 'Morzilla/5.0'})
html = BeautifulSoup(raw.text,'html.parser')
movies = html.select("dl.lst_dsc")

for m in movies:
    title = m.select_one('dt.tit>a')
    print(title.text)
    
    # 제목의 실제 주소값과 href의 속성값을 고려하여 아래와 같이 데이터 수집코드를 작성합니다.
    url = 'https://movie.naver.com'+title.attrs['href']

    # 새로운 링크에서 데이터를 수집합니다.
    raw_each = requests.get(url, headers={'User-Agent': 'Morzilla/5.0'})
    html_each = BeautifulSoup(raw_each.text, 'html.parser')
    reply = html_each.select("div.score_result li")
    for r in reply:
        score = r.select_one('div.star_score em').text
        reply = r.select_one('div.score_reple p').text
        print(score)
        print(reply)
````

### 파이썬 기본 라이브러리 사용
- 지금까지 사용한 requests, bs4, openpyxl과 같은 오픈소스 페키지는 별도의 설치가 필요한 파이썬 외부 라이브러리였습니다. 하지만 사람들이 많이 사용하기 때문에 별도의 설치가 필요 없는 파이썬에서 기본적으로 제공하는 내부 라이브러리도 있습니다.
- 여기선 urllib 라이브러리에 포함된 urlretrieve라는리기능을 사용합니다.

### 이미지 데이터 수집하기

#### 속성값 가져오기
```python
변수.attrs['속성이름']
```
- 선택한 코드에서 원하는 속성의 속성값을 저장합니다.
- poster에 저장되어있는 img 태그의 src 값을 저장하려면 poster.attrs['src']와 같이 쓸 수 있습니다.

#### 이미지 데이터 저장하기
src 값을 통해 이미지를 저장하기 위해 urlretrieve를 사용합니다.

#### urlretrive 사용 시 에러가 발생할 경우
```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```
코드를 통해 인터넷에 접속하여 파일을 다운로드 하는 방식이기 때문에 보안 상의 문제로 다운을 막아 놓는 경우가 있습니다. 위 코드를 추가하면 실행시 보안 정책을 일부 회피하여 다운로드를 가능케 합니다.

````python
import requests
from bs4 import BeautifulSoup
# urllib에서 urlretrive 가져오기
from urllib.request import urlretrieve
# 에러가 발생하는 경우 붙여 넣기
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

raw = requests.get("https://movie.naver.com/movie/running/current.nhn",
                  headers={'User-Agent': 'Morzilla/5.0'})
html = BeautifulSoup(raw.text,'html.parser')
movies = html.select("dl.lst_dsc")

for m in movies:
    title = m.select_one('dt.tit>a')
    print(title.text)
    
    url = 'https://movie.naver.com'+title.attrs['href']

    raw_each = requests.get(url, headers={'User-Agent': 'Morzilla/5.0'})
    html_each = BeautifulSoup(raw_each.text, 'html.parser')
    
    # 포스터의 src 속성값 찾기
    poster = html_each.select_one('div.mv_info_area div.poster img')
    poster_src = poster.attrs['src']
    # 이미지 데이터 저장
    # urlretrieve(poster_src, 'posters.png') --> 파일 덮어쓰기가 됨
    # poster 디렉토리 안에 제목.png라는 이름으로 이미지 저장
    urlretrieve(poster_src, 'poster/'+title.text[:2]+'.png')
````
